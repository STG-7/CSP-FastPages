## Pie Graph Example using NFL Teams

````python
import pandas as pd
import matplotlib.pyplot as plt

# create a new dataset related to the NFL
data = {'Team': ['Green Bay Packers', 'Dallas Cowboys', 'New England Patriots', 'Pittsburgh Steelers'],
        'Wins': [13, 10, 12, 9]}

df = pd.DataFrame(data)

# create a pie chart of the number of wins for each team
plt.pie(df['Wins'], labels=df['Team'], autopct='%1.1f%%')

# set the title
plt.title('Wins by NFL Team')

# show the chart
plt.show()
````

### Explanation of Code

This code creates a new dataset with the names of four NFL teams and the number of wins for each team. It then creates a pie chart of the number of wins for each team, using the pie() method instead of plot(kind='bar'). The autopct parameter formats the percentages displayed on each slice of the pie chart. Finally, it sets the title of the chart and displays it using show().

## Dot Graph Example using NBA Players (With .csv file)

````python
import pandas as pd
import matplotlib.pyplot as plt

# read the NBA dataset from a CSV file
nba_df = pd.read_csv('example.csv')

# create a scatter plot of points and rebounds
plt.scatter(nba_df['PTS'], nba_df['REB'], alpha=0.5)

# set the title and axis labels
plt.title('NBA Players: Points vs. Rebounds')
plt.xlabel('Points')
plt.ylabel('Rebounds')

# show the chart
plt.show()
````

### Explanation of Code

The example.csv file contains a dataset of NBA players with columns such as 'PTS' for points and 'REB' for rebounds, this code reads the dataset into a pandas DataFrame using read_csv(). It then creates a scatter plot of the players' points and rebounds using scatter(), with a transparency of 0.5 specified using the alpha parameter. Finally, it sets the title and axis labels of the chart and displays it using show().

# Questions with ANSWERS

1. What are the two primary data structures in pandas and how do they differ?

**The two primary data structures in pandas are Series and DataFrame. A Series is a one-dimensional labeled array that can hold any data type, while a DataFrame is a two-dimensional labeled data structure with columns of potentially different data types. DataFrames are more commonly used in pandas for data analysis as they can handle complex data manipulation tasks with ease, while Series are generally used as inputs or outputs for some operations.**

2. How do you read a CSV file into a pandas DataFrame?

**To read a CSV file into a pandas DataFrame, you can use the read_csv() method provided by pandas. First, import the pandas library. Then, use the read_csv() method to read the CSV file by providing the file path as a parameter. Finally, assign the result to a variable to store the DataFrame object. For example, import pandas as pd, df = pd.read_csv('file_path.csv') will read the CSV file from 'file_path.csv' and store the resulting DataFrame object in df.**

3. How do you select a single column from a pandas DataFrame?

**To select a single column from a pandas DataFrame, you can use the indexing operator [] with the name of the column as a string. For example, df['column_name'] will return a pandas Series object representing the selected column. Alternatively, you can also use the loc[] method to select a column by name: df.loc[:, 'column_name'].**

4. How do you filter rows in a pandas DataFrame based on a condition?

**To filter rows in a pandas DataFrame based on a condition, you can use boolean indexing. First, define the condition for filtering using logical operators and comparison operators, and apply it to the DataFrame using the indexing operator []. For example, df[df['column_name'] > 5] will return a DataFrame containing only the rows where the value in the column named 'column_name' is greater than 5. You can also use the query() method to apply a condition to a DataFrame: df.query('column_name > 5').**

5. How do you group rows in a pandas DataFrame by a particular column?

**To group rows in a pandas DataFrame by a particular column, you can use the groupby() method provided by pandas. First, select the column by which you want to group the rows using the indexing operator []. Then, call the groupby() method on the resulting Series or DataFrame object, passing the column name as a parameter. For example, df.groupby('column_name') will group the rows in the DataFrame df by the values in the column named 'column_name'.**

6. How do you aggregate data in a pandas DataFrame using functions like sum and mean?

**To aggregate data in a pandas DataFrame using functions like sum and mean, you can use the agg() method provided by pandas. First, group the rows in the DataFrame using the groupby() method. Then, call the agg() method on the resulting groupby object, passing the aggregation function(s) as a parameter. For example, df.groupby('column_name').agg({'column_to_aggregate': 'sum'}) will group the rows in the DataFrame df by the values in the column named 'column_name', and then sum the values in the column named 'column_to_aggregate'. You can also pass multiple aggregation functions to the agg() method, like df.groupby('column_name').agg({'column_to_aggregate': ['sum', 'mean']}).**

7. How do you handle missing values in a pandas DataFrame?

**To handle missing values in a pandas DataFrame, you can use the fillna() method provided by pandas. This method allows you to replace missing values with a specified value, such as a mean or median of the non-missing values in the same column. Alternatively, you can use the dropna() method to remove rows or columns with missing values, depending on the axis parameter. For example, df.fillna(df.mean()) will replace all missing values in the DataFrame df with the mean of the non-missing values in each column, while df.dropna(axis=0) will remove all rows with missing values.**

8. How do you merge two pandas DataFrames together?

**To merge two pandas DataFrames together, you can use the merge() method provided by pandas. First, define the two DataFrames you want to merge. Then, call the merge() method on one of the DataFrames, passing the other DataFrame as a parameter, and specify the join keys using the on parameter. For example, pd.merge(df1, df2, on='column_name') will merge the DataFrames df1 and df2 on the column named 'column_name'. You can also specify the type of join using the how parameter, like pd.merge(df1, df2, on='column_name', how='left') to perform a left join.**

9. How do you export a pandas DataFrame to a CSV file?

**To export a pandas DataFrame to a CSV file, you can use the to_csv() method provided by pandas. First, call the to_csv() method on the DataFrame you want to export, passing the file path as a parameter. You can also specify various options such as the delimiter, header, and index using the corresponding parameters. For example, df.to_csv('file_path.csv', index=False) will export the DataFrame df to the file 'file_path.csv' without including the index.**

10. What is the difference between a Series and a DataFrame in Pandas?

**In pandas, a Series is a one-dimensional labeled array that can hold any data type, while a DataFrame is a two-dimensional labeled data structure with columns of potentially different types. A DataFrame can be thought of as a collection of Series objects, where each column represents a Series. Series have a single column and no row labels, while DataFrames have both row and column labels.**
