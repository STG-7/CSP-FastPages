## HACK 2 Completion

```python
import numpy as np

# Create a random 3D array with shape (3, 4, 5)
a = np.random.randn(3, 4, 5)

# Use np.tensordot to compute the inner product of the last axis of a and the first axis of a
# This results in an array with shape (4, 5, 3, 3)
b = np.tensordot(a, a, axes=([0], [0]))

# Use np.transpose to swap the first and second axes of b
# This results in an array with shape (5, 4, 3, 3)
c = np.transpose(b, (1, 0, 2, 3))

print(c.shape)
```

## Explanation of Code

**This code uses the numpy library to create a random 3D array with shape (3, 4, 5) and then performs some array operations on it.**

**First, the np.tensordot function is used to compute the inner product of the last axis of a and the first axis of a. The axes argument specifies which axes to sum over for the inner product. In this case, we are summing over the first axis of the second a and the first axis of the first a. The resulting array b has shape (4, 5, 3, 3) because the first two axes are the result of the inner product.**

**Next, the np.transpose function is used to swap the first and second axes of b. This results in an array with shape (5, 4, 3, 3) because the first and second axes have been swapped.**

**Finally, the shape of the resulting array c is printed using the shape attribute.**

**In summary, this code creates a random 3D array, computes the inner product of the last and first axes, swaps the first and second axes, and prints the resulting shape. It is an example of how to use numpy functions to manipulate arrays in a powerful and flexible way.**

## Explanation of np.transpose() and np.tensordot() Functions

**np.transpose() is a function in NumPy that permutes the dimensions of an array. It takes as input an array and a tuple specifying the desired order of the dimensions. The output is an array with the same data but with the axes rearranged in the order specified by the tuple.**

**For example, np.transpose(a, (1, 0, 2)) swaps the first two dimensions of array a. If a has shape (3, 4, 5), the output will have shape (4, 3, 5).**

**np.tensordot() is a function in NumPy that computes the tensor dot product between two tensors. It takes as input two arrays and two lists of axes, one for each array. The axes are matched pairwise and summed over. The result is an array with shape equal to the concatenation of the non-contracted dimensions of the two input arrays.**

**For example, np.tensordot(a, b, axes=([1,2],[0,1])) computes the tensor dot product between a tensor a of shape (3,4,5) and a tensor b of shape (4,5,6). The output will have shape (3, 6), which corresponds to the dimensions not summed over.**

**Both np.transpose() and np.tensordot() are useful for manipulating and transforming multidimensional arrays in various ways. They are commonly used in scientific computing, machine learning, and data analysis.**
