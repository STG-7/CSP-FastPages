# HACK #3 - PROGRAMMING PARADIGMS

```python
import numpy as np
from PIL import Image

# Load the image
image = Image.open("../images/effeltower.jpg")

image.info['Title'] = 'Eiffel Tower'
image.info['Author'] = 'Shaurya GOel'
image.info['Year'] = '2023'

img_array = np.asarray(image)

binary_pixels = np.unpackbits(img_array, axis=-1)

hex_pixels = np.apply_along_axis(lambda x: hex(int(''.join(map(str, x)), 2))[2:].zfill(2), -1, binary_pixels)

red_img = np.copy(img_array)

red_img[:, :, 1] = 0
red_img[:, :, 2] = 0

red_image = Image.fromarray(red_img)

red_image.save('images/baltimore.jpg')

resized_image = red_image.resize((red_image.width // 4, red_image.height // 4))

display(resized_image)
```

**PARADIGM EXPLANATION**:

In the aforementioned hacks, I made an effort to use redscale (a different byte manipulation technique), print metadata (vital information about the image), and alter the image's size using scaling techniques. I used a photograph I took, Baltimore.jpg, and used the red img program to convert it to a redscale image by setting the green and blue channels to 0 (in "RGB"). Three printed pieces of image metadata were also added by me. This information contained the image's title, the photographer's name, and the year it was shot. When a jpg/jpeg file is compressed, additional metadata with the notation "jfif" is present. 

**TWO IMAGES REPRESENTING LOSSY AND LOSSLESS COMPRESSION**:

The lossy compression example will be the lassen-volcano.jpg image. This is owing to the fact that image file compression will undoubtedly result in a significantly smaller file size, but it will also result in a decrease in image quality because of the image's fine details and range of colors.
It's probably going to get fuzzier.On the other hand, the green-square-16.png image can be used as an illustration of lossless compression.
Because there is only one color in that image, the RGB pixel values will always be the same (without any intentional alteration of the image), the image quality will not change even though the image file size will reduce.
